## 导入数据

```python
fr = open('reviews.txt')
train_sent = []
for line in fr.readlines():
    curline = line.strip().split(' ')
    train_sent.append(curline)
fr.close()

fr = open('labels.txt')
train_labels = []
for line in fr.readlines():
    curline = line.strip()
    if 'positive' == curline:
        train_labels.append(1)
    else:
         train_labels.append(0)   
fr.close()
```

## 构建one-shot向量， 创建词集

```python
vocab = set()
for sent in train_sent:
    for word in sent:
        if(len(word)>0):
            vocab.add(word)
vocab = list(vocab)

word2index = {}
for i,word in enumerate(vocab):
    word2index[word]=i

input_dataset = list()
for sent in train_sent:
    sent_indices = list()
    for word in sent:
        try:
            sent_indices.append(word2index[word])
        except:
            ""
    input_dataset.append(list(set(sent_indices)))

target_dataset = list()
for label in train_labels:
    if label == 'positive\n':
        target_dataset.append(1)
    else:
        target_dataset.append(0)
```

## 加入嵌入层，将矩阵乘法变为矩阵选择对应行相加

### 数据初始化

```python
import numpy as np
np.random.seed(1)

def sigmod(x):
    return 1.0 / (1.0 + np.exp(-x))

alpha, iterations = (0.01, 2)
hidden_size = 100

weight_0_1 = 0.2 * np.random.random((len(vocab), hidden_size)) - 0.1
weight_1_2 = 0.2 * np.random.random((hidden_size, 1))  - 0.1
```

```python
correct, total = 0, 0
for iteration in range(iterations):
    for i in range(len(input_dataset) - 1000):
        x = input_dataset[i]
        y = target_dataset[i]
        # 选出输入节点相应的行 按列求和
        layer_1 = sigmod(np.sum(weight_0_1[x], axis=0))
        layer_2 = sigmod(np.dot(layer_1, weight_1_2))
        
        # 计算误差 
        layer_2_delta = layer_2 - y
        # 反向传播
        layer_1_delta = np.dot(layer_2_delta, weight_1_2.T)
        
        weight_1_2 -= alpha * layer_2_delta
        weight_0_1 -= alpha * layer_1_delta
```
